```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# loading the county-level census data
ctyData = pd.read_csv('nst-est2019-alldata.csv',
                      encoding = 'iso-8859-1')
```

```python
# Focusing on state totals (COUNTY==0)
stateData = ctyData[ctyData.STATE==0].copy()
ctyData.head()
```

```python
stateData.tail()
```

```python
# Sorting states in the order of population
stateData.sort_values(by='CENSUS2010POP',
                      ascending=False, inplace=True)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline



fig = plt.figure(figsize=(8,6))
fig.add_subplot(111)
b = plt.scatter(stateData['CENSUS2010POP'],stateData['REGION'],s=stateData['CENSUS2010POP']/1000,c=stateData['CENSUS2010POP']/100,label='Density')

cbar =fig.colorbar(b)

plt.xlabel('US,Brazil,India,Pakistan,Indonesia,China', fontsize=14)
plt.ylabel('Density', fontsize=14)

plt.tick_params(axis='both',labelsize=16)
cbar.ax.tick_params(labelsize=14)
cbar.set_label(label='Target Population', fontsize=16)





plt.tight_layout()
```

```python
stateNames = np.array(stateData.NAME)
statePop = np.array(stateData.CENSUS2010POP)
print('State\t\t\tPopulation')
for i,iState in enumerate(stateNames):
    print('%-20s' % iState + '\t%10d' % statePop[i])
```
